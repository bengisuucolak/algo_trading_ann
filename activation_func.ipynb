```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('fivethirtyeight')
```

```python
def binaryStep(x):
    ''' It returns '0' is the input is less then zero otherwise it returns one '''
    return np.heaviside(x,1)
```

```python
x = np.linspace(-5, 5)
plt.plot(x, binaryStep(x), color='r')
plt.axis('tight')
plt.title('AdÄ±m Fonksiyonu')
plt.show()
```

```python
def linear(x):
    ''' y = f(x) It returns the input as it is'''
    return x
```

```python
x = np.linspace(-5, 5)
plt.plot(x, linear(x),color='r')
plt.axis('tight')
plt.title('Lineer Fonksiyon')
plt.show()
```

```python

def sigmoid(x):
    ''' It returns 1/(1+exp(-x)). where the values lies between zero and one '''

    return 1/(1+np.exp(-x))
```

```python
x = np.linspace(-5, 5)
plt.plot(x, sigmoid(x),color='r')
plt.axis('tight')
plt.title('Sigmoid Fonksiyon')
plt.show()
```
